```python
%matplotlib inline
import cv2
import matplotlib
from matplotlib import pyplot as plt
image_path = "C:/Users/priya/Downloads/WhatsApp Image 2026-05-20 at 2.30.48 PM.jpeg"
image = cv2.imread(image_path)
resized_image = cv2.resize(image, (1900, 800))
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
plt.imshow(resized_image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
Gaussian = cv2.GaussianBlur(resized_image, (15, 15), 0)  
Gaussian_rgb = cv2.cvtColor(Gaussian, cv2.COLOR_BGR2RGB)  
plt.imshow(Gaussian_rgb)
plt.title('Gaussian Blurred Image')
plt.axis('off')
plt.show()
```

```python
Gaussian Blur is an image processing technique used to smooth an image and reduce noise. It works by applying a Gaussian filter, which averages pixel values based on a normal distribution.
In the context of cold steel roll coil image classification, the images may contain unwanted noise such as dust, lighting variations, or minor surface irregularities. These unwanted details can negatively affect the performance of the classification model.
```
