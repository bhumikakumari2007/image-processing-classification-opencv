```python
%matplotlib inline
import cv2
import matplotlib
from matplotlib import pyplot as plt
image_path = "C:/Users/priya/Downloads/WhatsApp Image 2026-05-20 at 2.30.48 PM.jpeg"
image = cv2.imread(image_path)
resized_image = cv2.resize(image, (1900, 800))
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
plt.imshow(resized_image_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
median = cv2.medianBlur(resized_image, 11)  
median_rgb = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)  

plt.imshow(median_rgb)
plt.title('Median Blurred Image')
plt.axis('off')
plt.show()
```

```python
# Median Blur.
```

```python
Median Blur is used to reduce noise in an image, especially salt-and-pepper noise. It works by replacing each pixel value with the median value of its neighboring pixels. In cold steel roll coil images, it helps remove small noise spots while preserving edges. This improves the clarity of important features for better image classification.
```
